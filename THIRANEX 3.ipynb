```python
# Creating historical sales dataset (more realistic)

data = {
    "Year": list(range(2015, 2023)),
    "Marketing_Spend": [50, 60, 65, 70, 80, 90, 100, 110],
    "Sales": [200, 230, 260, 280, 310, 350, 390, 420]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Features (independent variables)
X = df[["Year", "Marketing_Spend"]]

# Target (dependent variable)
y = df["Sales"]

print("✅ Data prepared")
```

```python
model = LinearRegression()
model.fit(X, y)

print("✅ Model trained successfully")
```

```python
# Predict on training data (since dataset is small)
y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MAE:", mae)
print("R2 Score:", r2)
```

```python
# Predict future sales

future_data = pd.DataFrame({
    "Year": [2023, 2024, 2025],
    "Marketing_Spend": [120, 130, 150]
})

future_predictions = model.predict(future_data)

for i in range(len(future_data)):
    print(f"Year {future_data['Year'][i]} → Predicted Sales: {int(future_predictions[i])}")
```
